<h2><font color='darkblue'>Parameter - hue</font></h2>

Laut Dokumentation ist der Paramter hue ein Parameter, der die Daten gruppiert. Diese Gruppierung wird farblich dargestellt.

Somit haben wir also einen Parameter mit dem wir eine weitere Information in dem Plot abbilden können.

Impliziert gleichzeitig auch, dass die Daten kategorisch und NICHT nummerisch sein dürfen.

Wenn du Daten die nummerisch sind, allerdings abbilden möchtest, dann wandelst du diese mit Hilfe von der cut() Funktion in kategorische Daten um.

```python
# hue - size

sns.scatterplot(x='total_bill',
                y='tip',
                data=df_tips,
                hue='size'            # gruppiert die Daten - Gruppierung wird farblich wiedergegeben
               
               )

```

Wir sehen, dass je dunkler die Farbe desto mehr Leute saßen an dem Tisch.

<h2><font color='darkblue'>Parameter - size</font></h2>

Laut Dokumentation ist size ein Paramter der die Daten gruppiert. Die Gruppierung wird anhand der unterschiedlichen 
Größe widergespiegelt. 

Das heißt wir haben einen weiteren Parameter mit dem wir eine weitere Information in dem Plot abbilden können.

```python
# size - time

sns.scatterplot(x='total_bill',
                y='tip',
                data=df_tips,
                hue='time',            # gruppiert die Daten - Gruppierung wird farblich wiedergegeben
                size='size',            # gruppiert die Daten - Gruppierung wird anhand der Größe wiedergegeben
                sizes=(1,199)
               )
```

Wir sehen, dass blaue Punkte Lunch und orange Punkte Dinner widerspiegeln. In Kombination mit der Anzahl der Personen,können wir sehen, ob große Gruppen eher zur Lunch oder zum Dinner sich zusammen finden.


<h2><font color='darkblue'>Parameter - style</font></h2>

Laut Dokumentation ist style ein Paramter der die Daten gruppiert. Die Gruppierung wird anhand von Markern widergespiegelt.

Folglich haben wir wieder einen Paramter mit dem wir eine weitere Information in dem Plot einbringen können. 

```python
# style - smoker

sns.scatterplot(x='total_bill',
                y='tip',
                data=df_tips,
                hue='time',            # gruppiert die Daten - Gruppierung wird farblich wiedergegeben
                size='size',           # gruppiert die Daten - Gruppierung wird anhand der Größe wiedergegeben
                style='smoker',        # gruppiert die Daten - Gruppierung anhand von Symbolen/Markern
                legend='auto'
               )
```

<div style="background-color:rgba(241, 158, 158, .5); margin:20px; padding:30px;">
<h5><font color='black'>Merkbox:</font></h5> 

Es existieren 3 Paramter mit denen du zusätzliche Informationen in deinem Plot einbringen kannst. Die Parameter gruppieren die Daten. Jeder Parameter für sich spiegelt die Gruppierung unterschiedlich wider.
    
        hue   --> anhand der Farbe
        size  --> anhand der Größe
        style --> anhand des Symbols/Markers
 
</div>

Die nachfolgenden Parametern behandeln das Customizing der drei Paramter.

<h2><font color='darkblue'>Parameter - hue - Customizing</font></h2>

<h3><font color='darkblue'>Parameter - hue - Customizing: Farbe</font></h3>

Laut Dokumentation kann die Farbe angepasst werden. Weiterhin ist schon die Übergabe der Parameter mit angegeben.

Voraussetzung ist, dass es sich um kategorische Daten handelt.

Wenn die Daten numerisch sind, dann kannst du diese mit der cut() Funktion in kategorische Daten umwandeln.
Wir können die Farben als:
                            
                            - string, 
                            - list, 
                            - dict, or
                            - matplotlib.colors.Colormap
                            
übergeben.

Wenn wir die Farben als dict übergeben, haben wir mehr "Kontrolle" über die Zuordnung der Farben. Bei einer Liste müssten wir die Reihenfolge der Daten beachten, wenn wir die Farben bestimmten Werten zuordnen möchten.

Wir probieren alle 4 Möglichkeiten einmal aus.

```python
# umwandeln der datentypen der spalte damit als keys verwendet werden können und die Variante dict möglich ist

df_tips_dic = df_tips.copy()
print(df_tips_dic.dtypes)
df_tips_dic['size'] = df_tips['size'].astype(str)
df_tips_dic.dtypes 
```

```python
# um auf cmap zugreifen zu können

import matplotlib 
import matplotlib.pyplot as plt 
```

```python
# hue - size

sns.scatterplot(x='total_bill',
                y='tip',
               # data=df_tips_dic,      # in kombi mit dic
                data=df_tips,
                hue='size',             # gruppiert die Daten - Gruppierung wird farblich wiedergegeben
               # palette='hsv'         # Variante string
               # palette=['purple','lightblue','lightgreen','red','orange','black'], # Variante list
                palette={1:'lightblue', 2:'lightgreen', 3:'red', 4:'orange', 5:'black', 6:'purple'},  # Variante dic
               # palette={'1':'lightblue', '2':'lightgreen', '3':'red', '4':'orange', '5':'black', '6':'purple'},  # Variante dic
               # palette=matplotlib.cm.Pastel1 # Variante colomap von matplotlib
               )
```

<h3><font color='darkblue'>Parameter - hue - Customizing: Anordnung</font></h3>

Mit dem Parameter hue_oder kann die Anordnung der Werte in der Legende angepasst werden. 

```python
# hue - size

sns.scatterplot(x='total_bill',
                y='tip',
                data=df_tips_dic,      # in kombi mit dic
                #data=df_tips,
                hue='size',             # gruppiert die Daten - Gruppierung wird farblich wiedergegeben
                palette='deep',         # Variante string
                hue_order=['6','4','2','3','1','5']  # Anordnung der Daten in der Legende
                #hue_order=[1,3,5,2,4,6]   # fuer df_tips
               )
```

<h3><font color='darkblue'>Parameter - hue - Customizing: Normalisieren</font></h3> 

Mit dem Parameter hue_norm können die Daten normalisiert werden.

```python
# hue - size

sns.scatterplot(x='total_bill',
                y='tip',
                #data=df_tips_dic,      # in kombi mit dic
                data=df_tips,
                hue='smoker', 
                size='size',# gruppiert die Daten - Gruppierung wird farblich wiedergegeben
                #palette='deep',         # Variante string
                #hue_order=['6','4','2','3','1','5'],  # anordnung der Daten in der Legende 
                #hue_norm=(0,6),         # setzt numerische kategorien voraus --> data anpassen--> Farbstärke
                sizes=(10,200),         # Variante string
                legend='full'
               )
```

Diese 2 Parameter für das Customizing findest du auch für die anderen 2 Parameter.

Wir machen einmal eine Zusammenfassung bzw. Gegenüberstellung.

                                                hue               size             style
       ----------------------------------|-----------------|-----------------|-----------------
       Gruppierung                       |    Farbe        |    Größe        |   Symbol/Marker
       Customizing - Gruppierung         |    palette      |    sizes        |   markers
       Customizing - Anordnung           |    hue_order    |    size_order   |   style_order
       Customizing - Daten normalisieren |    hue_norm     |    size_norm    |       n/a

<h2><font color='darkblue'>Parameter - ax</font></h2>

Wenn mehrere plots in subplots dargestellt werden sollen, dann wird die Anordnung des Plots zu dem Subplot über den Parameter ax gesteuert.


```python
import matplotlib 
import matplotlib.pyplot as plt 
```

```python
# figuren- und achsenobjekt erstellen

fig, ax = plt.subplots(nrows=2, ncols=2)
#fig, ax = plt.subplots(2, 2)
sns.scatterplot(x='total_bill',
                y='tip',
                #data=df_tips_dic,      # in kombi mit dic
                data=df_tips,
                hue='size',             # gruppiert die Daten - Gruppierung wird farblich wiedergegeben
                palette='deep',         # Variante string
                ax=ax[0][0]
               )

sns.scatterplot(x='total_bill', y='size', data=df_tips, ax=ax[1][0])
# Zugriff auf Achsenobjekt
#ax[1][1].set_title('Ich bin der Scatterplot', size=8, style='italic', family='serif')
```

<h2><font color='darkblue'>Titel einfügen</font></h2>

In der Funktion vom Scatterplot ist der Titel als Paramter nicht enthalten. Auch bei den kwargs ist der Titel nicht mit enthalten.

Aus diesem Grund greifen wir auf Matplotlib zurück. Dafür gibt es wieder verschiedene Möglichkeiten, um den Titel zu setzen.

```python
a = sns.scatterplot(x='total_bill',
                y='tip',
                #data=df_tips_dic,      # in kombi mit dic
                data=df_tips,
                hue='size',             # gruppiert die Daten - Gruppierung wird farblich wiedergegeben
                palette='deep'         # Variante string
                          )
a.set_title('Ich bin der Scatterplot')
print(type(a))
```
