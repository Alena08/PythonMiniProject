```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
#plt.scatter(tips["total_bill"], tips["tip"], alpha=0.6, s=80)
sns.scatterplot(x='total_bill',
                y='tip',
                data=tips,
                alpha=0.6, 
                s=80
)
```

```python
sns.histplot(tips["total_bill"],
             color="orange",
             bins=30)
```

```python
sns.boxplot(tips, x="sex", y='tip', color="lightgreen", linecolor="black")
```

```python
sns.barplot(tips, x="day", y="tip", color="purple", 
            errorbar=None, 
            linewidth=2.5, 
            edgecolor="black", 
            )
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5)) 

sns.histplot(tips["tip"],
             color="red",
             bins=30,
             ax=ax[0])
sns.scatterplot(tips, x="size", y="tip", alpha=0.6, 
                s=80, ax=ax[1],color="darkgreen" )
```

```python
sns.pairplot(tips,  x_vars=["total_bill", "tip", "size"], y_vars=["total_bill", "tip", "size"])
```
