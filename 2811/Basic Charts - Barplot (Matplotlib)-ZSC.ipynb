<h2><font color='darkblue'>Paramter - Barplot</font></h2>

Aus der Dokumentation geht hervor, dass der Barplot wie folgt aufgebaut ist:

matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

matplotlib.axes.Axes.bar(self, x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html

Wir verwenden die objektorientierte Arbeitsweise und werden auf dem Achsenobjekt die Funktion aufrufen und nicht pyplot verwenden.

Anhand der Paramter sehen wir schon einmal, dass wir die Paramter x UND height übergeben müssen.

In der Dokumentation sind die Paramter beschrieben.

Wir werden uns also zunächst:

                                1. non-default paramter
                                2. dann die default parameter 
                                3. und ein paar ausgewählte kwargs anschauen

```python
# benötigte bib importieren

import matplotlib 
import matplotlib.pyplot as plt 
import seaborn as sns # um einen Beispieldatensatz zu bekommen
```

```python
# datensatz aus seaborn laden

df_tips = sns.load_dataset('tips')
print(type(df_tips))
print(df_tips)
#print(df_tips.head(10))
```

```python
# daten für den barplot vorbereiten - wir wollen die Rechnungssumme anhand der Wochentage abbilden

df_tips_day = df_tips.groupby(['day'], as_index=False, observed=True).agg({'total_bill':'mean','smoker':'size'})
print(df_tips_day, type(df_tips_day))
df_tips_day
```

<h3><font color='darkblue'>1. Non-default Parameter</font></h3>
<h4><font color='darkblue'>Non-default Paramter x und height</font></h4>


```python
# Schritt für Schritt Parameter erkunden

# figuren- und achsenobjekt erstellen

fig, ax = plt.subplots() 

# barplot erstellen

ax.bar(x=df_tips_day['day'],
       height=df_tips_day['total_bill']
      
      )
```

<h3><font color='darkblue'>2. default Parameter</font></h3>
<h4><font color='darkblue'>default Parameter - width</font></h4>

Laut Doku gibt der Paramter width, die breite der plots an. Als Grundeinstellung sind 0.8 eingetragen.

Wir wollen einmal mit 0.4 ausprobieren, wie die barcharts aussehen.

```python
# figuren- und achsenobjekt erstellen

fig, ax = plt.subplots() 

# barplot erstellen

ax.bar(x=df_tips_day['day'],
       height=df_tips_day['total_bill'],
       width= 0.4                         # breite der balken
      
      )
```

<h4><font color='darkblue'>default Parameter - bottom</font></h4>

Laut Doku gibt der Parameter bottom an, an welchem Punkt die Barplots auf der y-achse beginnen.

Grundeinstellung ist die 0. 

Wir wollen hier einmal den Startpunkt bei 2.5 setzen.

```python
# figuren- und achsenobjekt erstellen

fig, ax = plt.subplots() 

# barplot erstellen

ax.bar(x=df_tips_day['day'],
       height=df_tips_day['total_bill'],   # kann durch Berechnung angepasst werden -2.5
       width=0.4,                          # breite der balken
       bottom=2.5                          # startpunkt der y-achse, keine automatische skalierung/anpassung der werte,
#       bottom=15                          # startpunkt der y-achse, keine automatische skalierung/anpassung der werte,
                                           # Überschreibung der Labels der y-Werte; alternativ: die Werte werden um den
                                           # Faktor vom botton (hier:2.5) verschoben      
      )
```

<h4><font color='darkblue'>default Parameter - align</font></h4>

Laut Doku gibt der Parameter align an, wie die Barplots auf der x-achse ausgerichtet sind.

Grundeinstellung ist zentriert. 

Wir wollen hier einmal edge uns anschauen.

```python
# figuren- und achsenobjekt erstellen

fig, ax = plt.subplots() 

# barplot erstellen

ax.bar(x=df_tips_day['day'],
       height=df_tips_day['total_bill'],
       width= -0.4,                       # - für linksseitig, positive Werte für rechtsseitig
       align='edge'                       # zeigt an, dass Balken nicht in der Mitte sondern am Rande 
                                          # ausgerichtet sind      
      )
```

<h3><font color='darkblue'>3. ausgewählte kwargs</font></h3>

In der Dokumentation findest du eine Auflistung an Parametern die du für das Customizing verwenden kannst.

Grundprinzipien sind:

                        - Farbeinstellungen
                        - Linienbreiten
                        - Beschriftungen (Label), Titel anpassen
                        - Texte in die Grafik bringen (Annotations)
                        ....
                        
                        
<h4><font color='darkblue'>kwargs - Farbe</font></h4> 

```python
# figuren- und achsenobjekt erstellen

fig, ax = plt.subplots() 

# barplot erstellen

ax.bar(x=df_tips_day['day'],
       height=df_tips_day['total_bill'],
       width=0.4,
       color=['b','orange','red','green'],    # Auswahl der farbe je barplot für den Inhalt
#       color=['blue','orange'],                  # Auswahl der farbe je barplot für den Inhalt
       edgecolor=['orange','red','green','blue'] # Auswahl der farbe je barplot für die Umrandung
      
      )
```

<h4><font color='darkblue'>kwargs - Linien</font></h4> 

```python
# figuren- und achsenobjekt erstellen

fig, ax = plt.subplots() 

# barplot erstellen

ax.bar(x=df_tips_day['day'],
       height=df_tips_day['total_bill'],
       width=0.4,
       color=['blue','orange','red','green'],      # auswahl der farbe je barplot für den inhalt
       edgecolor=['orange','red','green','blue'],  # auswahl der farbe je barplot für die Umrandung
       linewidth=[1,1,2,5],                        # Dicke der Linien
       linestyle=':'                              # Art der Linien
      )
```

<h4><font color='darkblue'>kwargs - Beschriftungen</font></h4>

```python
# figuren- und achsenobjekt erstellen
from time import sleep
fig, ax = plt.subplots() 

# barplot erstellen

ax.bar(x=df_tips_day['day'],
       height=df_tips_day['total_bill'],
       width=0.4,
       tick_label=['Donnerstag','Freitag','Samstag','Sonntag']
      )

# alles was nicht in der Funktion enthalten ist, kann mit der set() angepasst werden

# x_achsenbeschriftung

ax.set_xlabel('Wochentage')
# alternativ:

#ax.set(xlabel='weekday')

# y-achsenbeschriftung


ax.set_ylabel('durchschnittlicher Rechnungsbetrag')

# titel hinzufügen


ax.set_title('Rechnungsbeträge je Wochentag')

# labels der x-achse sollen leicht schräg sein

#ax.set_xticklabels(labels=['Donnerstag','Freitag','Samstag','Sonntag'], rotation='vertical')

ax.set_xticklabels(labels=['Donnerstag','Freitag','Samstag','Sonntag'], rotation=60)

plt.show() 
```

<h4><font color='darkblue'>kwargs - Texte (Annotations)</font></h4> 


```python
# figuren und achsenobjekt erstellen

fig, ax = plt.subplots() 

# barplot erstellen

my_barplot = ax.bar(x=df_tips_day['day'],
                   height=df_tips_day['total_bill'],
                   width=0.4,
                   tick_label=['Donnerstag','Freitag','Samstag','Sonntag']
                  )

# alles was nicht in der Funktion enthalten sit, kann mit der set() angepasst werden

# x_achsenbeschriftung

ax.set_xlabel('Wochentage')

# alternativ:

#ax.set(xlabel='weekday')

# y-achsenbeschriftung

ax.set_ylabel('durchschnittlicher Rechnungsbetrag')

# titel hinzufügen

ax.set_title('Rechnungsbeträge je Wochentag')

# labels der x-achse sollen leicht schräg sein

#ax.set_xticklabels(labels=['Donnerstag','Freitag','Samstag','Sonntag'], rotation='vertical')
ax.set_xticklabels(labels=['Donnerstag','Freitag','Samstag','Sonntag'], rotation=45)

# annotations setzen

for balken in my_barplot:
    height = balken.get_height()
    ax.annotate(f'{height:.2f}',
                    xy=(balken.get_x() + balken.get_width() / 2, height),
                    xytext=(0, 2),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

    
plt.show() 
```

Die Beträge passen nicht toll in die Grafik. Um diese Anpassen zu können, müssen wir die height anpassen, denn über die heigth werden die Werte auf den Balken gesteuert.

```python
# figuren und achsenobjekt erstellen

fig, ax = plt.subplots(figsize=(6,7)) 

# barplot erstellen

my_barplot= ax.bar(x=df_tips_day['day'],
                   height=df_tips_day['total_bill'],
                   width=0.4,
                   tick_label=['Donnerstag','Freitag','Samstag','Sonntag']
                  )

# alles was nicht in der Funktion enthalten sit, kann mit der set() angepasst werden

# x_achsenbeschriftung

ax.set_xlabel('Wochentage')

# alternativ:

#ax.set(xlabel='weekday')

# y-achsenbeschriftung

ax.set_ylabel('durchschnittlicher Rechnungsbetrag')

# titel hinzufügen

ax.set_title('Rechnungsbeträge je Wochentag')

# labels der x-achse sollen leicht schräg sein

#ax.set_xticklabels(labels=['Donnerstag','Freitag','Samstag','Sonntag'], rotation='vertical')
ax.set_xticklabels(labels=['Donnerstag','Freitag','Samstag','Sonntag'], rotation=45)

# annotations setzen

for balken in my_barplot:
    height = balken.get_height()
#    ax.annotate(text='$ {:.1f}'.format(height),
    ax.annotate(text=f'{height:.2f} €',
                xy=(balken.get_x() + balken.get_width() / 2, height),
                xytext=(0, 3),                                         # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
    
plt.show() 
```
