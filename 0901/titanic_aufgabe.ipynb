```python
#insert bestimmen, wo die Spalte stehen soll (hier an Index 7).
neu_list = df["sibsp"] + df["parch"] + 1
df.insert(7,'family_size',neu_list)
```

```python
df.columns
```

```python
#Das Alter in Kategorien einteilen
df['age_cut'] = pd.cut(df['age'],3,labels=["kids", "jung", 'old'])
```

```python
df.loc[:,['age','age_cut']].head()
```

```python
#Statistiken pro Gruppe erstellen
df.groupby('age_cut').agg({'fare':'mean', 'survived':'sum'})
```

```python
#Eine Übersichtstabelle der Überlebensraten erstellen 
pivot = df.pivot_table(
    values='survived',   # Was berechnet werden soll
    index='pclass',  # Die Zeilen    
    columns='sex',   # Die Spalten   
    aggfunc=['mean',"sum"] 
)
pivot
```

```python
categories = ["Allein", "Familie"]
einteilung = [0, 1, 15]
df["familien_size_cut"] = pd.cut(df["family_size"], bins=einteilung, labels=categories)
```
