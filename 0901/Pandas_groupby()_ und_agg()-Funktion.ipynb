```python
# benötigte bib importieren

import seaborn as sns 
import pandas as pd 
import numpy as np 
```

```python
sns.get_dataset_names()
```

```python
# Beispieldatensatz

df_tips = sns.load_dataset('tips')
print(df_tips.head())
#df_tips.head(+200)

```

Um die Frage: wieviele Männer und Frauen besuchen unser Restaurant? zu beantworten, muss die spalte geschlecht zunächst nach männern und frauen gruppiert werden. danach kann die anzahl ermittelt werden.

```python
# Wie viele Frauen und Männer essen bei uns im restaurant?

# group by - erst gruppieren und danach eine Aggregationsfunktion aufrufen

df_tips.groupby('sex').size()  # Anzahl der Elemente
```

Wenn du mehrere Werte berechnen möchtest, gibst du im groupby deine spalte an die gruppiert werden soll. die zu berechnenden Werte gibst du in der aggregationsfunktion an.

Frage: An welchen Tagen wird wie viel Trinkgeld im Schnitt gegeben und wie viele Leute kommen an den Tagen?

Vorüberlegung: 
              Zu Gruppieren ist nach den Tagen. Die Berechnung erfolgt für das Trinkgeld und die Tage

```python
# An welchen Tagen wird wieviel Trinkgeld gegeben und wieviele Leute waren im restaurant?

df_tips.groupby('day').agg({'tip': np.mean, 'day': np.size})
```

```python
# alternativ:'median' = pandas Methoden

#df_tips.groupby('day').agg({'tip': ['mean', np.median, np.min], 'day': np.size})
df_tips.groupby('day').agg({'tip': ['mean', 'median'], 'day': np.size})
```

```python
# Pandas: Gruppieren nach 2 Kriterien 

df_tips_gruppiert = df_tips.groupby(['day', 'smoker']).agg({'tip': [np.size, np.mean]})
#df_tips_gruppiert.groupby(level='smoker','day')
df_tips_gruppiert.loc[:,'tip']
```

```python
# wie komme ich an smoker?

df_tips_gruppiert.loc[:,'tip']

# spalte exisitert nicht?
```

```python
df_tips_gruppiert.columns 
```

```python
# also smoker und day sind ein index --> gibt unterschiedliche Wege damit umzugehen

# eine Möglichkeit --> as_index=False setzen

df_tips_gruppiert2 = df_tips.groupby(['smoker', 'day'],as_index=False).agg({'tip': [np.size, np.mean]})
df_tips_gruppiert2 
```

<h2><font color='darkblue'>Unterschied zu Pivot_table()</font></h2>

```python
# beispiel datensatz

df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",

                         "bar", "bar", "bar", "bar"],

                   "B": ["one", "one", "one", "two", "two",

                         "one", "one", "two", "two"],

                   "C": ["small", "large", "large", "small",

                         "small", "large", "small", "small",

                         "large"],

                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],

                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})

df
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],columns=['C'], aggfunc=np.sum)

table 
```

```python
table.columns 
```

```python
# nachbauen mit groupby und aggregate

table_groupby = df.groupby(['A','B','C']).agg({'D':[np.sum]})
table_groupby 
```
