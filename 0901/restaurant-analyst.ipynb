```python
import pandas as pd 
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# an welchen Wochentagen (day) das Restaurant geöffnet hat 
df['day'].unique()
```

```python
#wie viele Tische an jedem dieser Tage bedient wurden 
df['day'].value_counts()
```

```python
#neue Spalte 'tip_pct' 
tip_prozent = df['tip']/df['total_bill']
df.insert(2,'tip_pct', tip_prozent)
```

```python
df.head()
```

```python
# Kategorien bilden
categories = ["Günstig", "Normal", "Teuer"]
einteilung = [0,15,25,100]
df['bill_class'] = pd.cut(df['total_bill'], bins=einteilung, labels=categories)
```

```python
df['bill_class'].head()
```

```python
df.groupby('bill_class').agg({"tip":"mean","size":"max","day":"count"})
```

```python
pivot = df.pivot_table(
    values='tip_pct',   # Was berechnet werden soll
    index='day',  # Die Zeilen    
    columns='time',   # Die Spalten   
    aggfunc='mean' 
)
```
