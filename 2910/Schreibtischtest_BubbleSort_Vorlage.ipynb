```python
# Funktion BubbleSort(a) 
# Es wird eine Liste (a) bergeben und dann sortiert

def BubbleSort(a) :
   #for i in range(len(a)-1, 0, -1):
    n = len(a)
    i = n-1
    while i > 0:
        j = 1
        while j <= i:
            if a[j-1] > a[j]:
                tausche(a, j)
            j+=1
        i-=1

# tausche(a, j) tauscht, bei Bedarf, die beiden Listenelemente

def tausche(a, j):
        temp = a[j-1]
        a[j-1]=a[j]
        a[j]=temp


###############    
### M A I N ###
###############

a = [ 7, 4, 12, 1, 2]

BubbleSort(a)
print('#######################################')
print('Ergebnis:             ',a, sep='')
```
