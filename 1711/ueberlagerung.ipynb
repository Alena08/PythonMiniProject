```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 500)
#y = np.sin(x)
fig, ax = plt.subplots(figsize=(10, 5))  
print(fig, ax)
ax.plot(x,np.sin(x), label="Welle A")
ax.plot(x,0.5*np.sin(5*x), label="Welle B")
ax.plot(x,np.sin(x) + 0.5*np.sin(5*x), label="berlagerung")
ax.grid()
ax.set_xlabel("x label")                        # Add an x-label to the axes.
ax.set_ylabel("y label")                        # Add a y-label to the axes.
ax.set_title("berlagerung zweier Wellen")  # Add a title to the axes.
ax.legend()   
```
