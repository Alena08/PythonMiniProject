Stelle dir vor, du möchtest eine Rechteckfunktion (oder Pulswelle) visualisieren, die
folgende Bedingungen erfüllt:

1.Sie existiert im Intervall von x = -10 bis x = 10

Im Intervall von x = -5 bis x = 5 soll die Funktion den Wert 1 haben.

Außerhalb dieses Intervalls (also von x = -10 bis -5 und von x = 5 bis 10) soll die Funktion
den Wert 0 haben.



```python
import numpy as np
import matplotlib.pyplot as plt

plt.title("Rechteckfunktion")
plt.xlabel("X")
plt.ylabel("Y")
x = np.linspace(-10, 10, 201)

y = np.zeros(x.shape)

#index = np.where((x >= -5) & (x <= 5))
#print(index[0][0])
y[(x >= -5) & (x <= 5)] = 1 #(-5 <= x <= 5)
plt.axhline(0, c="green", linewith = 1)

plt.plot(x, y)
```
