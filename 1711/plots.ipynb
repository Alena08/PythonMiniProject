```python
plt.title("Rechteck")
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter([-2,-2,2,2],[-2,2,-2,2])
plt.annotate("A",(-1.9,-1.9))
plt.annotate("B",(1.8,-1.9))
plt.annotate("C",(1.8,1.8))
plt.annotate("D",(-1.9,1.8))
plt.plot([-2,2],[-2,-2])
plt.text(-1.9, 0, "a", color='orange', size=14.5)
plt.plot([-2,-2],[-2,2])
plt.text(0, -1.9, "b", color='blue', size=14.5)
plt.plot([-2,2],[2,2])
plt.text(0, 1.8, "d", color='green', size=14.5)
plt.plot([2,2],[-2,2])
plt.text(1.8, 0, "c", color='red', size=14.5)
```

```python

plt.xlabel('x')
plt.ylabel('y') 
plt.arrow(0, 0, 3, 3, head_width=0.1, linestyle='-', length_includes_head=True)
plt.arrow(0, 0, -3, 2, head_width=0.2, ls=':', length_includes_head=True)
plt.arrow(0, 0, -4, -1, head_width=0.3, length_includes_head=True)
plt.text(3.1, 3.1, 'v1')
plt.text(-3.3, 2.4, 'v2')
plt.text(-4.3, -1.6, 'v3')
plt.grid()
 
 
plt.axis([-5, 5, -5, 5])
plt.axhline(Linewidth=2)
plt.axvline(Linewidth=2)
 


 
```
